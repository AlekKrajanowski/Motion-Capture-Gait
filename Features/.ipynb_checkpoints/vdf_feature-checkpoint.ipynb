```python
from pyomeca import Markers
import math
import matplotlib.pyplot as plt
from statistics import mean
import numpy as np
from scipy.stats import skew 

data_path = "data/data.c3d"

def vdf_method():
    
    channels = ['RFHD','RWRA','RSHO','RANK','LANK','RHEE','LHEE']
    markers_body = Markers.from_c3d(data_path, usecols=channels)
    
    channels_sym = ['CentreOfMass']
    markers_sym = Markers.from_c3d(data_path, usecols=channels_sym)
    
    #vd1 HEIGHT
    vd1 = []
    vd1_y = 0
    #vd2 WRIST RIGHT
    vd2 = []
    vd2_y = 0
    #vd3 SHOULDER RIGHT
    vd3 = []
    vd3_y = 0
    #vd4 ANKLE RIGHT
    vd4 = []
    vd4_y = 0
    #vd5 ANKLE LEFT
    vd5 = []
    vd5_y = 0
    #vd6 DIST. FEET LEFT AND RIGHT
    vd6 = []
    vd6_y = 0
    
    #variables
    meanV = []
    stdV = []
    VDF = [] 
    vdf = []
    time = []
    average = []
    i = 0
    
    while i < 485:
    
        vd1_y = markers_body[1][0][i]
        vd1.append(vd1_y)
        
        vd2_y = markers_body[1][1][i]
        vd2.append(vd2_y)
        
        vd3_y = markers_body[1][2][i]
        vd3.append(vd3_y)
        
        vd4_y = markers_body[1][3][i]
        vd4.append(vd4_y)
        
        vd5_y = markers_body[1][4][i]
        vd5.append(vd5_y)
        
        vd6_y = (0.5 * (markers_body[1][5][i] - markers_body[1][6][i]) * markers_sym[1][0][i])
        vd6.append(vd6_y)
        
        time.append(i)
        
        average = (vd1[i] + vd2[i] + vd3[i] + vd4[i] + vd5[i] + vd6[i]) / 6
        vdf = [vd1[i], vd2[i], vd3[i], vd4[i], vd5[i], vd6[i]]
        
        meanV.append(average)
        stdV.append(np.std(vdf))
        VDF.append([meanV, stdV])
        
        i += 1
        
        return VDF, time
        
#     fig, axs = plt.subplots(1, 3)
#     axs[0].plot(time, meanV, 'tab:red')
#     axs[0].set_title('MeanV')
#     axs[1].plot(time, stdV, 'tab:blue')
#     axs[1].set_title('StdV')
#     axs[2].plot(time, VDF[0][0], 'tab:red')
#     axs[2].plot(time, VDF[1][1], 'tab:blue')
#     axs[2].set_title('VDF')
    
#     for ax in axs.flat:
#         ax.set(xlabel='time(h)', ylabel='variable')
        
#     for ax in axs.flat:
#         ax.label_outer()
    
vdf_method()
        
```
